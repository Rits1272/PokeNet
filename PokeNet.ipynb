```python
import keras
from keras.models import Sequential
from keras.layers import Conv2D, Dense, Flatten, MaxPooling2D
from keras.optimizers import Adam
from sklearn import preprocessing
import pandas as pd
import random
import numpy as np
from keras.preprocessing import image
from keras.utils.np_utils import to_categorical
```

```python
data = pd.read_csv("train.csv")
```

```python
data['NameOfPokemon'].unique()
```

```python
X = data['ImageId']
y = data['NameOfPokemon']

label_encoder = preprocessing.LabelEncoder()
y = label_encoder.fit_transform(y) # Hot Encoding the data
y = to_categorical(y)
print(y)
```

```python
# Training and Testing Data
X_train = X[61:305]
X_test = X[:61]
data_dic = {}
j=0

for i in range(296,498):
    path = './Images/'+str(i-295)+'.jpg'
    data_dic[path] = y[j]
    j += 1

for i in range(203,305):
    path = './Images/'+str(i)+'.jpg'
    data_dic[path] = y[j]
    j+=1

d = data_dic # Backup
# Reshuffling the data
keys = list(data_dic.keys())
random.shuffle(keys)
shuffled_data_list = [(key, data_dic[key]) for key in keys]
print(shuffled_data_list[0])
print(shuffled_data_list[0][0])
print(shuffled_data_list[0][1])
len(shuffled_data_list)
```

```python
X_train = []
X_test = []
y_train = []
y_test = []

for i in range(245):
    img = image.load_img(shuffled_data_list[i][0], target_size=(75,75,3))
    img = image.img_to_array(img)
    img = img/255
    X_train.append(img)
    y_train.append(shuffled_data_list[i][1])

for i in range(245,304):
    img = image.load_img(shuffled_data_list[i][0], target_size=(75,75,3))
    img = image.img_to_array(img)
    img = img/255
    X_test.append(img)
    y_test.append(shuffled_data_list[i][1])
num_pixels =75*75*3
X_test = np.array(X_test)
X_train = np.array(X_train)
print(X_train.shape)

y_train = np.array(y_train)
y_test = np.array(y_test)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
def model():
    model = Sequential()
    model.add(Conv2D(30, (3,3), input_shape=X_train.shape[1:], activation='relu'))
    model.add(Conv2D(20, (3,3), activation='relu'))
    model.add(MaxPooling2D(pool_size=(2,2)))
    model.add(Conv2D(10,(3,3),activation='relu'))
    model.add(Flatten())
    model.add(Dense(64, activation='relu'))
    model.add(Dense(3, activation='softmax'))
    model.compile(loss='categorical_crossentropy',optimizer='adam', metrics=['accuracy'])
    model.fit(X_train, y_train, batch_size=20, validation_split=0.1, epochs=35)
    return model
```

```python
model = model()
```

```python
score = model.evaluate(X_test, y_test, verbose=2)
print(str(score[1]*100)+"%")
```
